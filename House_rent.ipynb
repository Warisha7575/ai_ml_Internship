```python
import pandas as pd
data = pd.read_csv(r'C:\Users\warisha khan\Desktop\training\House_Rent_Dataset.csv')
data.head()
```

```python
y = data["Rent"]
x= data.drop(['Rent','Posted On', 'Floor', 'Area Locality'], axis=1)
```

```python
import category_encoders as ce
encoder = ce.LeaveOneOutEncoder()
X= encoder.fit_transform(x,y)
print(X)
```

```python
#min-max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(X)
print(x)
z= pd.DataFrame(x, columns = X.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
r = model.fit(x_train, y_train)
r = model.predict(x_train[1:2])
print(r)
```

```python
new_house = pd.DataFrame({
    'BHK': [2],
    'Size': [1200],
    'Area Type': ['Super Area'],
    'City': ['Delhi'],
    'Furnishing Status': ['Semi-Furnished'],
    'Tenant Preferred': ['Family'],
    'Bathroom': [2],
    'Point of Contact': ['Contact Owner']
})
new_encoded = encoder.transform(new_house)
new_scaled = scaler.transform(new_encoded)
prediction = model.predict(new_scaled)
print("Predicted Rent for the new house:", prediction[0])
```
